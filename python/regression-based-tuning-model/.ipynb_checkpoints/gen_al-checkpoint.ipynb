```python
from operator import itemgetter
import random, string, sys
```

```python
def fitness(word, password):
    score = 0
    if len(word)!=len(password):
        print('Length mismatch of password')
        score = -1
    else:
        for i in range(len(word)):
            if word[i]==password[i]:
                score += 1
    return score/len(word)*100
```

```python
def createWord(length):
    word = []
    for i in range(length):
        word.append(random.choice(string.ascii_lowercase))
    return "".join(word)
```

```python
def initialize_population(no_of_population,password):
    population = []
    for i in range(no_of_population):
        population.append(createWord(len(password)))
    return population
```

```python
def getFitness(population,password):
    population_fitness = {}
    for i in population:
        population_fitness[i] = fitness(i,password)
    population_fitness = sorted(population_fitness.items(),key=lambda x: x[1],reverse=True)        
    return population_fitness
```

```python
def crossover(word1,word2):
    child = ''
    for i in range(len(word1)):
        if random.random() > 0.5:
            child += word1[i]
        else:
            child += word2[i]
    return child
```

```python
def mutation(word):
    mutated_word = ''
    for i in range(len(word)):
        random_letter = random.choice(string.ascii_lowercase)
        if 100*random.random() > 50:
            mutated_word += random_letter
        else:
            mutated_word += word[i]
    return mutated_word
```

```python
def selectBreeder(population,no_of_best_individuals, no_of_lucky_individuals):
    breeders = []
    for i in range(no_of_best_individuals):
        breeders.append(population[i][0])
    for i in range(lucky_individuals):
        breeders.append(random.choice(population)[0])
    random.shuffle(breeders)
    return breeders
```

```python
def nextGeneration(population, no_of_best_individuals, no_of_lucky_individuals, password, no_of_children):
    next_generation = []
    score = getFitness(population,password)
    breeders = selectBreeder(score,no_of_best_individuals, no_of_lucky_individuals)
    l = len(breeders)
    print('Number of breeders: '+str(l))
    for i in range(int(l/2)):
        for j in range(no_of_children):
            next_generation.append(crossover(breeders[i],breeders[l-1-i]))
    for i in range(len(next_generation)):
        if random.random()*100 > 50:
            next_generation[i] = mutation(next_generation[i])
    return next_generation,score[0][0],score[0][1]
```

```python
no_of_population = 100
best_individuals = 20
lucky_individuals = 20
children = 5
password = 'alfnwonfoqwefn'
generation = 100

if no_of_population != (best_individuals+lucky_individuals)/2*children:
    print('Incompatible individual/children selection')
    sys.exit()

population = initialize_population(no_of_population,password)
for i in range(generation):
    print("Generation: "+str(i))
    [population,best_individual,best_score] = nextGeneration(population,best_individuals,lucky_individuals,password,children)
    print('Best score and individual is:'+str(best_score)+' '+best_individual)
    if best_score == 100:
        print('Metahumans have been evolved. Aborting rest of task.')
        break
```
